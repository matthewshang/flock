```python
import jax
import jax.numpy as jnp
import numpy as np
import math
import matplotlib.pyplot as plt

def random_unit_vector(key: jnp.ndarray) -> jnp.ndarray:
    key, subkey = jax.random.split(key)
    theta = jax.random.uniform(key, (1,)) * 2 * math.pi
    z = jax.random.uniform(subkey, (1,)) * 2 - 1
    r = jnp.sqrt(1 - z**2)
    return jnp.hstack([r * jnp.cos(theta), r * jnp.sin(theta), z])

key = jax.random.PRNGKey(0)

ax = plt.figure().add_subplot(projection='3d')


# Plot scatterplot data (20 2D points per colour) on the x and z axes.
colors = ('r', 'g', 'b', 'k')

vecs = jax.vmap(random_unit_vector)(jax.random.split(key, 1000))
# By using zdir='y', the y value of these points is fixed to the zs value 0
# and the (x, y) points are plotted on the x and z axes.
ax.scatter(vecs[:, 0], vecs[:, 1], vecs[:, 2], c='r')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```

```python
from flax import struct
import jax
import jax.numpy as jnp

@struct.dataclass
class Foo:
    x: float
    y: float

def init_one(x, y):
    return Foo(x, y)

def init_many(x, y):
    return jax.vmap(init_one)(x, y)

init_many(jnp.arange(5), jnp.arange(5))
```
